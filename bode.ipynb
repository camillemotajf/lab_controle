```python
import numpy as np
import matplotlib.pyplot as plt

# Frequências
frequencies = [0.1, 1, 10]  # Hz
omega_c = 100  # Frequência de corte (Hz)

# Módulos (em dB) e fases (em graus) correspondentes às frequências
modulos = [25.57, 25.42, 13.3]  # Exemplo de valores de módulo (em dB)
fases = [-7.13, -10.46, -43.29] # Exemplo de valores de fase (em graus)

# Convertendo fases para radianos
fases_rad = np.radians(fases)

# Interpolação para outras frequências
new_frequencies = np.logspace(-1, 2, 1000)  # Frequências de 0.1 Hz a 100 Hz
new_omegas = 2 * np.pi * new_frequencies  # Frequências angulares
new_modulos = []
new_fases = []

# Interpolação para o módulo
for omega in new_omegas:
    new_modulo = modulos[0] - 20 * np.log10(np.sqrt(1 + (omega / omega_c)**2))
    new_modulos.append(new_modulo)

# Interpolação para a fase
for omega in new_omegas:
    new_fase = -np.degrees(np.arctan(omega / omega_c))
    new_fases.append(new_fase)

# Plotando a assíntota do diagrama de Bode interpolada
plt.figure()

# Plotando o módulo em dB
plt.subplot(2, 1, 1)
plt.semilogx(new_frequencies, new_modulos, 'b-')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Módulo (dB)')
plt.grid(True)

# Plotando a fase em graus
plt.subplot(2, 1, 2)
plt.semilogx(new_frequencies, new_fases, 'r-')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Fase (graus)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
