```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('dados_senoide_10hz.csv')
df2 = pd.read_csv('dados_rpm_10hz.csv')
```

```python
print(len(df1))
```

```python
x1 = df1['tempo']
senoide1 = df1['tensao']
x2 = df2['tempo'].iloc[:8000]
senoide2 = df2['amostra_ajustada'].iloc[:8000]
```

```python
# Calcular os deslocamentos para centralizar as senoides
# Calcular o deslocamento para centralizar a senoide1
deslocamento1 = -np.mean(senoide1)
senoide1_centralizada = senoide1 + deslocamento1

# Criar a figura e o eixo principal
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotar a senoide original no eixo principal
ax1.plot(x2, senoide2, 'b-', label='Senoide 2')
ax1.set_xlabel('x')
ax1.set_ylabel('Amplitude Senoide 2', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Criar um segundo eixo y que compartilha o mesmo eixo x
ax2 = ax1.twinx()

# Plotar a senoide centralizada no segundo eixo y
ax2.plot(x1, senoide1_centralizada, 'r-', label='Senoide 1 Centralizada')
ax2.set_ylabel('Amplitude Senoide 1', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Adicionar uma linha de referncia no eixo y=0 para a senoide centralizada
ax2.axhline(0, color='black', linewidth=0.5, linestyle='--')

# Adicionar ttulos e legenda
plt.title('Senoides com Escalas Diferentes e Centralizao')
fig.tight_layout()  # Para ajustar o layout e evitar sobreposio

plt.show()
```
